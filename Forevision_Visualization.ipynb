```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
df=pd.read_csv('Sales Data.csv')
```

```
df.head()
```

```
print(df.info())
print(df.describe())
```

```
features=['Sales (in Cr)','Profit (in Cr)']
df[features].hist(figsize=(10,4))
```

```
df[features].plot(kind='density',subplots=True,layout=(1,2),sharex=False,figsize=(12,4))
```

```
sns.distplot(df['Sales (in Cr)'])
```

```
sns.boxplot(x='Profit (in Cr)',data=df)
```

```
sns.countplot(x="Remark",data=df)
```

```
plt.scatter(x='Sales (in Cr)',y='Profit (in Cr)',data=df)
```

```
sns.jointplot(x='Sales (in Cr)',y='Profit (in Cr)',data=df,kind='kde')
```

```
sns.lmplot('Sales (in Cr)','Profit (in Cr)',data=df,hue='Remark',fit_reg=False)
```

```
df['Remark'].value_counts().plot.pie(shadow=True,figsize=(10,8))
plt.show()
```
